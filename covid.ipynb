# General Topic Search Function

```python
def tweet_topics(search, n_tweets, n_topics):
    """
    Given a search term, the number of tweets to pull, and the number of topics, returns popular words assoiated with each topic. 
    
    Parameters
    search: string. Search term for the tweets.
    n_tweets: int. Number of tweets to pull.
    n_topics: int. Number of topics to break the tweets into. 

    """
    a = twint.Config()
    a.Search = search
    a.Lang = 'en'
    a.Limit = n_tweets
    a.Pandas = True
    a.format = '{tweet}\n'

    twint.run.Search(a)
    
    display.clear_output()

    df = twint.storage.panda.Tweets_df
    df = df[df['language'] == 'en']
    
    #custom_stop_words = stop_words + ' ' + search
    
    vec = CountVectorizer(stop_words = frozenset(stop_words.split()))
    words = vec.fit_transform(df['tweet'])
    
    nmf = NMF(n_topics)
    topic = nmf.fit_transform(words)
    
    features = vec.get_feature_names()
    t = nmf.components_.argsort(axis=1)[:,-1:-7:-1]
    topic_words = [[features[e] for e in l] for l in t]
    
    return topic_words
```

```python
import gensim
from sklearn.preprocessing import StandardScaler
```

```python
def word2vec_tweets(query, n_tweets):
    a = twint.Config()
    a.Search = query
    a.Lang = 'en'
    a.Limit = n_tweets
    a.Pandas = True

    twint.run.Search(a)
    
    print('search finished')
    
    display.clear_output()

    df = twint.storage.panda.Tweets_df
    df = df[df['language'] == 'en']
    
    docs = [tweet for tweet in df['tweet']]
    
    all_words = [word.lower().strip().split() for word in docs]
    
    slimmed_docs = []
    
    for tweet in all_words:
        for word in tweet:
            slim_tweet = [word for word in tweet if word not in stop_words]
            
        slimmed_docs.append(slim_tweet)
        
    model = gensim.models.Word2Vec(slimmed_docs, size=100, window=5, min_count=1, workers=2, sg=1)
    
    return list(model.wv.vocab.items())[:5]
```

# General Sentiment Analyzer Function

```python
def tweet_sentiment(search, n_tweets):
    a = twint.Config()
    a.Search = search
    a.Lang = 'en'
    a.Limit = n_tweets
    a.Pandas = True

    twint.run.Search(a)
    
    display.clear_output()

    df = twint.storage.panda.Tweets_df
    df = df[df['language'] == 'en']
    
    df['polarity'] = df['tweet'].apply(lambda x: TextBlob(x).sentiment[0])
    df['subjectivity'] = df['tweet'].apply(lambda x: TextBlob(x).sentiment[1])
    
    print('mean polarity:', df['polarity'].mean())
    print('mean subjectivity:', df['subjectivity'].mean())
```

# General Polarity and Subjectivity Analyzer Functions

```python
def polarity(search):
    a = twint.Config()
    a.Search = search
    a.Lang = 'en'
    a.Limit = 10000
    a.Pandas = True

    twint.run.Search(a)
    
    display.clear_output()

    df = twint.storage.panda.Tweets_df
    df = df[df['language'] == 'en']
    
    df['polarity'] = df['tweet'].apply(lambda x: TextBlob(x).sentiment[0])
    
    return df['polarity'].mean()
```

```python
def subjectivity(search):
    a = twint.Config()
    a.Search = search
    a.Lang = 'en'
    a.Limit = 10000
    a.Pandas = True

    twint.run.Search(a)
    
    display.clear_output()

    df = twint.storage.panda.Tweets_df
    df = df[df['language'] == 'en']
    
    df['subjectivity'] = df['tweet'].apply(lambda x: TextBlob(x).sentiment[1])
    
    return df['subjectivity'].mean()
```
